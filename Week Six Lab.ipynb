Today we need to first add a 2nd dimension to the while loop of last weeks lab. 
Here is the data from the last weeks lab with a changing x value. Recall that this models an object being dropped straight down. 

```python
import numpy as np
import matplotlib.pyplot as plt
dt=0.001
g=9.8
m=0.0032
C=0.5
p=1.2
A=0.002
t=[0]
y=[2]
x=[0]
vx=[0]
vy=[0]
ax=[0]
ay=[-g]
while(y[-1]>0):
    t=t+[t[-1]+dt]
    y=y+[y[-1]+vy[-1]*dt]
    x=x+[x[-1]+vx[-1]*dt]
    vy=vy+[vy[-1]+ay[-1]*dt]
    vx=vx+[vx[-1]+ax[-1]*dt]
    ay=ay+[(-m*g+0.5*p*C*A*(((vx[-1]**2)+(vy[-1]**2))**0.5)*(vy[-1]))/m]
    ax=ax+[((0.5*p*C*A*((vx[-1]**2)+(vy[-1]**2))**0.5)*(vx[-1]))/m]
plt.plot(x,y)
plt.title('X versus Y')
plt.xlabel('x (m)')
plt.ylabel('y (m)')
```

## Launch angle: 30 degrees

Height of Launch: 0.265 m

The Initial Velocity of the Steel Ball is 4.9 m/s (x: 4.24 m/s, y: 2.45 m/s)

The Mass of The Steel 0.0666 kg

Radius: 0.000254 m, Cross Sectional Area: 0.000000203


The Initial Velocity of the Plastic Ball 4.8 m/s (x: 4.15, y: 2.4)

The Mass of The Plastic is: 0.01296 kg

Radius 0.000243 m, Cross Sectional Area: 0.000000186

2.57 m for plastic actual distance from barrel.

```python
#Steel Ball 

dt=0.01
g=9.8
m=0.0666
C=0.5
p=1.2
A=0.000000203
t=[0]
y=[0.265]
x=[0]
vx=[4.24]
vy=[2.45]
ax=[0]
ay=[-g]
while(y[-1]>0):
    t=t+[t[-1]+dt]
    y=y+[y[-1]+vy[-1]*dt]
    x=x+[x[-1]+vx[-1]*dt]
    vy=vy+[vy[-1]+ay[-1]*dt]
    vx=vx+[vx[-1]+ax[-1]*dt]
    ay=ay+[(-m*g+0.5*p*C*A*(((vx[-1]**2)+(vy[-1]**2))**0.5)*(-vy[-1]))/m]
    ax=ax+[((0.5*p*C*A*((vx[-1]**2)+(vy[-1]**2))**0.5)*(-vx[-1]))/m]
plt.plot(x,y)
plt.title('X versus Y')
plt.xlabel('x (m)')
plt.ylabel('y (m)')
```

Above is a projection. For the trajectory of the metal ball. 

According to this it should land a little past 2.5 meters from the barrel. The measured value is 2.62 meters. That seems pretty good to me. 

Lets move on to the plastic ball: 

```python
#Plastic Ball 

dt=0.01
g=9.8
m=0.01296
C=0.5
p=1.2
A=0.000000186
t=[0]
y=[0.265]
x=[0]
vx=[4.15]
vy=[2.4]
ax=[0]
ay=[-g]
while(y[-1]>0):
    t=t+[t[-1]+dt]
    y=y+[y[-1]+vy[-1]*dt]
    x=x+[x[-1]+vx[-1]*dt]
    vy=vy+[vy[-1]+ay[-1]*dt]
    vx=vx+[vx[-1]+ax[-1]*dt]
    ay=ay+[(-m*g+0.5*p*C*A*(((vx[-1]**2)+(vy[-1]**2))**0.5)*(-vy[-1]))/m]
    ax=ax+[((0.5*p*C*A*((vx[-1]**2)+(vy[-1]**2))**0.5)*(-vx[-1]))/m]
plt.plot(x,y)
plt.title('X versus Y')
plt.xlabel('x (m)')
plt.ylabel('y (m)')
```

This projection indicates that the ball should land approximately 2.5 meters from the barrel. The actual distance was 2.57 m. This seems pretty darn acuarate.
